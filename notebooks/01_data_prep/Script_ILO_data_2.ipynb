```python
import os
import requests
import pandas as pd
import json
import time
```

```python
# Carpeta donde guardaremos los CSV
output_folder = "/Users/veliaalaminos/Desktop/DAFT_July2025/Projects/Final Project/2nd Approach/data/raw/ILO_data"
os.makedirs(output_folder, exist_ok=True)

# Lista de países
countries = {"Brazil": "BRA_A", 
             "China": 'CHN_A', 
             "Spain": 'ESP_A', 
             "Finland": 'FIN_A', 
             "France": 'FRA_A', 
             "United Kingdom": "GBR_A", 
             "Indonesia": "IDN_A",
             "India": "IND_A", 
             "Ireland": "IRL_A", 
             "Iraq": "IRQ_A", 
             "Libya": "LBY_A", 
             "Marocco": "MAR_A", 
             "Mexico": "MEX_A", 
             "Poland": "POL_A", 
             "Portugal": "PRT_A", 
             "Saudi Arabia": "SAU_A",
             "Sudan": "SDN_A",
             "Sweden": "SWE_A",
             "Thailand": "THA_A",
             "South Africa": "ZAF_A",
             "Russia": "RUS_A"
}


# Indicadores clave seleccionados
indicators = [
    "SDG_0552_NOC_RT",      # Employment by occupation
    "SDG_T552_NOC_RT",       # Total employment by occupation
    "SDG_0831_SEX_ECO_RT",   # Employment by economic activity
    "SDG_0851_SEX_OCU_NB",   # Employees by occupation and sex
    "SDG_0852_SEX_AGE_RT",   # Employment by age and sex
    "EAP_TEAP_SEX_AGE_EDU_NB", # Education level
    "EAP_DWAP_SEX_AGE_RT",     # Working age population by sex and age
    "EMP_TEMP_SEX_AGE_NB",     # Temporary employment
    "UNE_TUNE_SEX_AGE_NB"      # Unemployment by age and sex
]

# Construimos la URL base de ILO
base_url = "https://rplumber.ilo.org/data/ref_area/"

# Rango de años
time_from = 2013
time_to = 2023

# Formato de descarga
file_format = ".csv"


for country_name, iso in countries.items():
    try:
        # Construimos la URL
        url = f"{base_url}?id={iso}&indicator={'+'.join(indicators)}&timefrom={time_from}&timeto={time_to}&type=label&format={file_format}"
        
        # Hacemos la petición
        response = requests.get(url)
        response.raise_for_status()
        
        # Guardamos el CSV
        filepath = os.path.join(output_folder, f"{iso}.csv")
        with open(filepath, "wb") as f:
            f.write(response.content)
        
        print(f"✅ Descargado {country_name} ({iso})")
        
    except Exception as e:
        print(f"❌ Excepción al descargar {country_name} ({iso}): {e}")
```
