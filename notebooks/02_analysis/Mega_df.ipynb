```python
import pandas as pd
import os
import glob
```

```python
# Carpeta con los CSVs
csv_folder = "/Users/veliaalaminos/Desktop/DAFT_July2025/Projects/Final Project/2nd Approach/data/raw/ILO_data/"

# Lista todos los CSV
csv_files = glob.glob(os.path.join(csv_folder, "*.csv"))

# Lista para guardar cada df
dfs = []

for file in csv_files:
    try:
        df = pd.read_csv(file)
        
        # Añadir columna con el país (extraemos del nombre del archivo)
        country_iso = os.path.basename(file).replace(".csv", "")
        df['ISO_A'] = country_iso
        
        dfs.append(df)
    except Exception as e:
        print(f"Error leyendo {file}: {e}")

# Unir todos los df en uno solo
mega_df = pd.concat(dfs, ignore_index=True)
```
